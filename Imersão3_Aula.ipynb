###Desafio_08: Resumo das informa√ß√µes coletadas dos dados.

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace = True)
##dados.rename = Alterar o nome de uma linha/coluna.
##inplace = True == Alterar dados raiz.
```

```python
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
##Lista criada com todos os 5 primeiros compostos indexados. 
```

```python
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
## Tudo que estiver contido na lista cod_compostos retornar√° para o dataframe.
## "@" utilizamos o arroba para solicitar a inclus√£o do c√≥digo em python na query.
## Query √© como se um pequeno c√≥digo estivesse rodando dentro do "seu c√≥digo". 
## in = contido
```

```python
import seaborn as sns
sns.set() ## Chama a configura√ß√£o padr√£o do Seaborn.
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 5))
ax = sns.countplot(x = 'composto',data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show()

##biblioteca seaborn importada, sns.countplot(x = 'composto',data=dados.query('composto in @cod_compostos')) utilizado para gerar o gr√°fico. 
##.set_title = Incluir t√≠tulo
##Neste caso armazenamos o gr√°fico em uma vari√°vel tida como padr√£o na comunidade como "ax", deixando o c√≥digo mais simples me assim seguimos com o 'set.title'.
##plt.show() retira informa√ß√µes que n√£o utilizamos.
```
