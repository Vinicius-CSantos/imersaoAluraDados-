###Desafio_08: Resumo das informações coletadas dos dados.

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace = True)
##dados.rename = Alterar o nome de uma linha/coluna.
##inplace = True == Alterar dados raiz.
```

```python
dados.head()
```

```python
cod_compostos = dados['composto'].value_counts().index[0:5]
##Lista criada com todos os 5 primeiros compostos indexados. 
```

```python
cod_compostos
```

```python
dados.query('composto in @cod_compostos')
## Tudo que estiver contido na lista cod_compostos retornará para o dataframe.
## "@" utilizamos o arroba para solicitar a inclusão do código em python na query.
## Query é como se um pequeno código estivesse rodando dentro do "seu código". 
## in = contido
```

```python
import seaborn as sns
sns.set() ## Chama a configuração padrão do Seaborn.
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 5))
ax = sns.countplot(x = 'composto',data=dados.query('composto in @cod_compostos'))
ax.set_title('Top 5 Compostos')
plt.show()

##biblioteca seaborn importada, sns.countplot(x = 'composto',data=dados.query('composto in @cod_compostos')) utilizado para gerar o gráfico. 
##.set_title = Incluir título
##Neste caso armazenamos o gráfico em uma variável tida como padrão na comunidade como "ax", deixando o código mais simples me assim seguimos com o 'set.title'.
##plt.show() retira informações que não utilizamos.
```
