```python
##Desafio 5:
# correla√ß√µes entre g's vista em aula

corr_gen = dados.loc[:,'g0':'g50'].corr() # calcula correla√ß√£o entre vari√°veis. [-1,1] => altamente correlacionadas, ~=0 -> pouco correlacionadas
corr_gen.head()
```

```python

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_gen, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_gen, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
gxc = pd.concat([dados.loc[:,'g0':'g50'],dados.loc[:,'c0':'c50']],axis=1).corr() # calcula correla√ß√£o entre vari√°veis. [-1,1] => altamente correlacionadas, ~=0 -> pouco correlacionadas
gxc.head()
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(gxc, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(gxc, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(gxc, dtype=bool)) # cria uma m√°scara nos dados do dataframe (gxc) selecionando o tri√¢ngulo superior da matriz partindo da diagonal central (k=0). O par√¢metro dtype=bool transforma os valores em booleanos

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9)) # define os objetos gr√°ficos e o tamanho das figuras

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True) # cria um custom colormap para ser usado no heatmap

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(gxc, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}) # plota o gr√°fico do tipo heatmap utilizando o custom cmap e o mask criados anteriormente. O par√¢metro mask usa o tri√¢ngulo superior para esconder os valores da matriz.
```
