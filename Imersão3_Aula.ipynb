```python
##DESAFIO 1
##Ordenado de maneira decrescente.
ordem.decresc = dados.query('composto in @cod_compostos')['composto'].value_counts().index
ordem.decresc
##Ordenado de maneira crescente.
## ordem.cresc = dados.query('composto in @cod_compostos')['composto'].value_counts().sort_values().index
```

```python
##Desafio 2

plt.figure(figsize=(6, 5))
ax = sns.countplot(x = 'composto',data=dados.query('composto in @cod_compostos'),order=ordem.decresc,palette='Paired')
ax.set_title('Top 5 Compostos usados', fontsize=16, weight='bold')
ax.set_ylabel('Quantidade',fontsize=12, weight='bold')
ax.set_xlabel('Composto',fontsize=12, weight='bold')
plt.show()
```

```python
#Desafio 03

plt.figure(figsize=(10,7))
ax = sns.histplot(data=dados,x='g19',kde=True,element="step",stat='density')
ax.set_title('Densidade gene g-19',fontsize=20, weight='bold')
ax.set_ylabel('Frequência Relativa',fontsize=15, weight='bold')
ax.set_xlabel('Gene g-19',fontsize=12, weight='bold')
plt.show()


```
